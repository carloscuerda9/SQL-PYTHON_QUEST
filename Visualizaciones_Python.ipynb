```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = pd.read_csv('Tabla de totales (P.1).csv')
```

```python
data1
```

```python
data1.shape
```

```python
data1 = data1.sort_values(by='record_count', ascending=False)

# Crear un grfico de barras con Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='table_name', y='record_count', data=data1, palette='viridis')
plt.title('Total de Productos por Marca (Ordenado de Mayor a Menor)')
plt.xlabel('Marca')
plt.ylabel('Nmero de Productos')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
data2 = pd.read_csv('Tabla de totales por marca (P.2).csv')
```

```python
data2.head(15)
```

```python
data2.shape
```

```python
data2['brand_name'] = data2['brand_name'].str.replace('"', '')

# Rellenar los valores NaN con 0 para evitar problemas en la visualizacin
data2['total_products'] = data2['total_products'].fillna(0)

data2 = data2.sort_values(by='total_products', ascending=False)

# Crear un grfico de barras con Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='brand_name', y='total_products', data=data2, palette='viridis')
plt.title('Total de Productos por Marca')
plt.xlabel('Marca')
plt.ylabel('Nmero de Productos')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
data3 = pd.read_csv('Tabla categorias-n.productos (P.3).csv')
```

```python
data3.head(15)
```

```python
data3.shape
```

```python
# Ordenar el DataFrame por la columna 'total_products' de mayor a menor
data3 = data3.sort_values(by='total_products', ascending=False)

# Crear un grfico de barras con Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='category_name', y='total_products', data=data3, palette='viridis')
plt.title('Nmero de Productos por Categora (Ordenado de Mayor a Menor)')
plt.xlabel('Categora')
plt.ylabel('Nmero de Productos')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
data4 = pd.read_csv('Promedio de productos por pedido (P.4).csv')
```

```python
data4.head(15)
```

```python
data4.shape
```

```python
average_products_per_order = 2.9238

# Redondear el valor a 2 decimales
average_products_per_order_rounded = round(average_products_per_order, 2)

# Crear un cuadro de texto con el valor redondeado
plt.figure(figsize=(6, 4))
plt.text(0.5, 0.5, f'Promedio de productos por pedido:{average_products_per_order_rounded}', 
         ha='center', va='center', fontsize=14, bbox=dict(facecolor='lightgrey', alpha=0.5))
plt.axis('off') 
plt.title('Promedio de Productos por Pedido')
plt.show()
```
